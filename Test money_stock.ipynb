```python
import interest_rate as ir
import money_stock as ms
import plotly.express as px
import plotly.graph_objects as go
```

```python
def ts_plot(df,date_range,keys,names,title):

    fig = go.Figure()
    
    for k,n in zip(keys,names):
        
        fig.add_trace(
            go.Scatter(
                x=df.index, 
                y=df[k], 
                name=n
            )
        )

    fig.update_layout(
        title_text=title,
        xaxis_rangeslider_visible=True
    )
    
    fig.show()
```

```python
# load the data vendor
get_ms = ms.vendor()
get_ir = ir.vendor()
```

```python
# Load H6 table
H6 = get_ms.FRED_H6()
H6.info()
```

```python
df = H6['1995':].drop(columns='dep_savings').copy()
ts_plot(
    df,
    df.index,
    df.columns,
    df.columns,
    'Money base'
)
```

```python
# Load H4 table
H4_avg = get_ms.FRED_H4_weeklyavg()
H4_wed = get_ms.FRED_H4_asofwed()
H4_avg.info()
H4_wed.info()
```

```python
df = H4_avg['2008':].copy()
ts_plot(
    df,
    df.index,
    ['fedrrp_mmf','fedrrp_fo','res_bal','tga'],
    ['Fed RRP w/ MMF','Fed RRP w/ Foreign Officials','Reserve balances','TGA'],
    'Factors Absorbing Reserves'
)
```

```python
# Load H8 data
H8 = get_ms.H8()
H8.columns
```

```python
df = H8.unstack().swaplevel(axis=1)['Large']
keys = ['Cash assets',
         'Treasuries',
         'Total fed funds sold and reverse repo',
         'Loans to nondepository financial institutions']
ts_plot(df, df.index, keys, keys, 'Cash investments by commercial banks')
```
