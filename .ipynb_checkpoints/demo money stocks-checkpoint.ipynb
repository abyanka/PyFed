```python
!pip install -r requirements.txt
!pip install plotly
import PyFed.money_stock as ms
import plotly.express as px
import plotly.graph_objects as go
```

```python
def ts_plot(df,date_range,keys,names,title):

    fig = go.Figure()
    
    for k,n in zip(keys,names):
        
        fig.add_trace(
            go.Scatter(
                x=df.index, 
                y=df[k], 
                name=n
            )
        )

    fig.update_layout(
        title_text=title,
        xaxis_rangeslider_visible=True
    )
    
    fig.show()
```

```python
get_ms = ms.vendor() # load the data vendor
```

# Fed Board H6 table

```python
H6 = get_ms.FRED_H6(); # Load H6 table 
H6.info()
```

```python
df = H6['1995':].drop(columns='dep_savings').copy()
ts_plot(
    df,
    df.index,
    df.columns,
    ['Money market funds: retail','Money market funds: institutional',
    'Small time deposits','Demand deposits','Other checkable deposits'],
    'Money base'
)
```

# Fed Board H4 table

```python
H4_avg = get_ms.FRED_H4_weeklyavg() # Load H4 table: weekly averages
H4_wed = get_ms.FRED_H4_asofwed() # Load H4 table: weekly as of wednesday
H4_avg.info()
H4_wed.info()
```

```python
df = H4_avg['2008':].copy()
ts_plot(
    df,
    df.index,
    ['fedrrp_mmf','fedrrp_fo','res_bal','tga'],
    ['Fed RRP w/ MMF','Fed RRP w/ Foreign Officials','Reserve balances','TGA'],
    'Factors Absorbing Reserves'
)
```
